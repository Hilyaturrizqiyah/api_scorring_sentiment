```python
import pandas as pd
import re
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from nltk.tag import CRFTagger
from nltk.tag import pos_tag
```

```python
df = pd.read_csv("crawlingcut.csv")
```

```python
df.drop_duplicates(subset='text', inplace=True)
```

```python
def str_to_lower(text):
    return text.lower()


df['username_lower'] = df['username'].apply(lambda x: str_to_lower(x))
df['text_lower'] = df['text'].apply(lambda x: str_to_lower(x))
df.head()
```

```python
def remove_punct(text):
    text = re.sub(r'[^a-zA-Z]', ' ', str(text))
    text = re.sub(r'\b\w{1,2}\b', '', text)
    text = re.sub(r'\s\s+', ' ', text)
    text = re.sub('([#])|([^a-zA-Z])', ' ', text)
    return text


df['twitter_account'] = df['username'].apply(lambda x: remove_punct(x))
df['ISI'] = df['text_lower'].apply(lambda x: remove_punct(x))
df.head()
```

```python
def tokenization(text):
    # text = nltk.tokenize.word_tokenize(text)
    # return(text)
    text = re.split(r'\W+', text)
    return text


df['tokenization'] = df['ISI'].apply(lambda x: tokenization(x))
df.head()
```

```python
df['tokenization'][0]
```

```python
stopword = pd.read_csv('stopword.csv')
stopwords = [x for x in stopword['ada']]
stopwords
```

```python
def remove_stopword(text):
    text = [word for word in text if word not in stopwords]
    return text

df['remove_stopword'] = df['tokenization'].apply(lambda x: remove_stopword(x))
df.head()
```

```python
normalisasi_kata = pd.read_csv("singkatan.csv")

normalisasi_kata_dict = {}

for row in normalisasi_kata.iterrows():
    if row[0] not in normalisasi_kata_dict:
        normalisasi_kata_dict[row[0]] = row[1]


def normalisasi_term(document):
    return [normalisasi_kata_dict[term]
            if term in normalisasi_kata_dict
            else term for term in document]


df['normalisasi_tweet'] = df['tokenization'].apply(normalisasi_term)

df['normalisasi_tweet'].head(0)
```

```python
df['normalisasi_tweet'].head(0)
```

```python
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()


# stemmed
def stemmed_wrapper(term):
    return stemmer.stem(term)


term_dict = {}

for document in df['normalisasi_tweet']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '

print(len(term_dict))
print("------------------------")

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term, ":", term_dict[term])

print(term_dict)
print("------------------------")


# apply stemmed term to dataframe
def get_stemmed_term(document):
    return [term_dict[term] for term in document]


df['tweet_tokens_stemmed'] = df['normalisasi_tweet'].apply(get_stemmed_term)
print(df['tweet_tokens_stemmed'])
```

```python
df.to_csv('result.csv', index=False)
```

```python
from nltk.tag import CRFTagger

ct = CRFTagger()
ct.set_model_file('all_indo_man_tag_corpus_model.crf.tagger')
k = ct.tag_sents(df.tokenization)

final = []
for i in k:
    word = []
    for n in i:
        tag = list(n)
        # a, b = tag
        # print(tag)
        if ((tag[1] == "VB") or (tag[1] == "JJ") or (tag[1] == "RB")):
            word.append(tag[0])
    one_tweet = " ".join(word)
    if (len(one_tweet) > 0):
        final.append(one_tweet)
dict1 = {'POS_Tweet': final}
db1 = pd.DataFrame(dict1)
filename = "postagging.csv"
db1.to_csv(filename)
```
